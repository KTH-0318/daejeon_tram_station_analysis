```python
# make_geojson_simple.py
import pandas as pd, json

CSV_PATH = r"C:\Users\THKIM\Desktop\3-2 프로젝트\스마트시티 문제해결 프로젝트\2호선_위_경도.csv"          # 입력 CSV
GEOJSON_OUT = r"C:\Users\THKIM\Desktop\3-2 프로젝트\스마트시티 문제해결 프로젝트\daejeon_line2.geojson"  # 출력 GeoJSON

df = pd.read_csv(CSV_PATH, encoding="utf-8-sig")  # CP949면 encoding="cp949"

# 숫자 변환(문자 섞였을 때 대비)
df["위도"] = pd.to_numeric(df["위도"], errors="coerce")
df["경도"] = pd.to_numeric(df["경도"], errors="coerce")

# 좌표 없는 행 제외
df = df.dropna(subset=["위도", "경도"])

features = []
for _, r in df.iterrows():
    props = {
        "정거장ID": r.get("정거장ID"),
        "정거장명": r.get("정거장명")
    }
    features.append({
        "type": "Feature",
        "properties": props,
        "geometry": {"type":"Point","coordinates":[float(r["경도"]), float(r["위도"])]}  # [경도, 위도]
    })

geojson = {"type":"FeatureCollection","features":features}
with open(GEOJSON_OUT, "w", encoding="utf-8") as f:
    json.dump(geojson, f, ensure_ascii=False, indent=2)

print(f"Saved -> {GEOJSON_OUT}  (features: {len(features)})")
```
